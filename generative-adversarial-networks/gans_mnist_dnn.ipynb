```
import tensorflow as tf
import numpy as np

import matplotlib.pyplot as plt

!nvidia-smi
```

```
def preprocessing(x, shape):
    x = np.asarray(x, dtype=np.float32).reshape(shape)
    x = x / 255.0
    # rerange to [-1, 1]
    x = 2. * x - 1.
    return x


(x_train, _), (_, _) = tf.keras.datasets.mnist.load_data()

img_shape = (28, 28, 1)
z_size = 128
x_train = preprocessing(x_train, shape=(-1, *img_shape))


def noise(batch_size, z_size):
    return tf.random.normal((batch_size, z_size), dtype=tf.float32)


class GANs_DNN(object):
    def __init__(self, img_shape, z_size, k=1, dropout=0.0):
        self.img_shape = img_shape
        self.pixels = np.prod(img_shape)
        self.z_size = z_size
        self.dropout = dropout
        self.generator = self.create_generator()
        self.discriminator = self.create_discriminator()

        self.global_step = tf.Variable(0, name='global_step', trainable=False)

        self.d_optimizer = tf.keras.optimizers.Adam(0.0001)
        self.g_optimizer = tf.keras.optimizers.Adam(0.0001)
        self.k = k

        self.acc_data = tf.keras.metrics.BinaryAccuracy('data_accuracy')
        self.acc_generate = tf.keras.metrics.BinaryAccuracy('generate_accuracy')

    @staticmethod
    def mean_log(values):
        return tf.reduce_mean(tf.math.log(values))

    def create_generator(self):
        x = tf.keras.Input(shape=(self.z_size,))
        h = tf.keras.layers.Dense(
            256, 'relu', 
            kernel_initializer=tf.initializers.TruncatedNormal(stddev=0.1))(x)
        h = tf.keras.layers.Dropout(self.dropout)(h)
        h = tf.keras.layers.Dense(
            256, 'relu', 
            kernel_initializer=tf.initializers.TruncatedNormal(stddev=0.1))(h)
        h = tf.keras.layers.Dropout(self.dropout)(h)
        h = tf.keras.layers.Dense(
            self.pixels, 'tanh', 
            kernel_initializer=tf.initializers.TruncatedNormal(stddev=0.1))(h)
        h = tf.reshape(h, shape=(-1, *self.img_shape))
        model = tf.keras.Model(inputs=x, outputs=h, name='Generator')
        # model.summary()
        return model

    def create_discriminator(self):
        x = tf.keras.Input(shape=self.img_shape)
        h = tf.reshape(x, shape=(-1, self.pixels))
        h = tf.keras.layers.Dense(
            256, 'relu', 
            kernel_initializer=tf.initializers.TruncatedNormal(stddev=0.1))(h)
        h = tf.keras.layers.Dense(
            256, 'relu', 
            kernel_initializer=tf.initializers.TruncatedNormal(stddev=0.1))(h)
        h = tf.keras.layers.Dense(
            1, 'sigmoid', 
            kernel_initializer=tf.initializers.TruncatedNormal(stddev=0.1))(h)
        model = tf.keras.Model(inputs=x, outputs=h, name='Discriminator')
        # model.summary()
        return model

    def train(self, dataset, epochs, batch_size, shuffle=True):
        if shuffle:
            np.random.seed(7)
            np.random.shuffle(dataset)
        
        valid_noise = noise(24, self.z_size)
        for epoch in range(1, epochs + 1):
            for i in range(0, len(dataset), batch_size):
                x_data = dataset[i:i + batch_size]
                x_noise = noise(np.shape(x_data)[0], self.z_size)
                d_loss, g_loss = self.train_step(x_data, x_noise)

                step = self.global_step.numpy()
                if step % 500 == 0:
                    print('eopch:%d step:%d d_loss:%.4f g_loss:%.4f acc_data:%.4f acc_gene:%.4f' % (
                        epoch, step, d_loss.numpy(), g_loss.numpy(),
                        self.acc_data.result().numpy(),
                        self.acc_generate.result().numpy()))
                    self.acc_data.reset_states()
                    self.acc_generate.reset_states()

            if epoch % 20 == 0:
                print('\nepoch %d, generated images:' % epoch)
                valid_generated = self.generator(valid_noise)
                valid_generated = tf.squeeze(valid_generated)
                self.show_images(valid_generated)

    @tf.function(
        input_signature=[tf.TensorSpec((None, *img_shape)), tf.TensorSpec((None, z_size))])
    def train_step(self, x_data, x_noise):
        """training once step"""
        with tf.GradientTape(persistent=True) as tape:
            x_generated = self.generator(x_noise, training=True)
            y_generated = self.discriminator(x_generated, training=True)
            y_data = self.discriminator(x_data)
            d_loss = - self.mean_log(y_data) - self.mean_log(1.0 - y_generated)
            g_loss = - self.mean_log(y_generated)

        # update generator parameters
        if self.global_step % self.k == 0:
            g_grads = tape.gradient(g_loss, self.generator.trainable_variables)
            g_grads = [tf.clip_by_value(grad, -5.0, 5.0) for grad in g_grads]
            self.g_optimizer.apply_gradients(zip(g_grads, self.generator.trainable_variables))

        # update discriminator parameters
        d_grads = tape.gradient(d_loss, self.discriminator.trainable_variables)
        d_grads = [tf.clip_by_value(grad, -5.0, 5.0) for grad in d_grads]
        self.d_optimizer.apply_gradients(zip(d_grads, self.discriminator.trainable_variables))

        self.global_step.assign_add(1)
        self.acc_data(tf.ones_like(y_data, tf.int32), y_data)
        self.acc_generate(tf.ones_like(y_generated, tf.int32), y_generated)

        return d_loss, g_loss

    @staticmethod
    def show_images(images):
        plt.figure(figsize=(9, 6))
        for i, image in enumerate(images, 1):
            plt.subplot(4, 6, i)
            plt.imshow(image, cmap='gray')
            plt.axis('off')
        plt.show()


model = GANs_DNN(img_shape=img_shape, z_size=z_size, k=1, dropout=0.3)
model.train(x_train, epochs=500, batch_size=256)
# model.discriminator(x_train[:2], training=True)
```
